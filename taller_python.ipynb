### Ahora vamos a ver Bucles: while, for, break, continue
El while y for permiten crear bucles que ejecutan una sección de código repetidas veces. Por otro lado el continue y el break permiten realizar alteraciones sobre el bucle.

#### El ciclo while ejecuta su sección de código mientras se cumpla una determinada condición.

```
x = 0
while x < 8:
    print(x)
    x += 1
```

#### El ciclo for permite iterar clases iterables, ejecutando la sección de código tantas veces como elementos tiene el iterable.

```
for i in range(8):
    print(i)
```

#### El continue salta hasta el final del bloque, dejando sin ejecutar lo restante, pero continúa en la siguiente iteración. (del ciclo for)



```
for i in range(8):
    if i == 3:
        continue
    print(i)
```

#### Por último, el break rompe la ejecución del bucle, saliendo del mismo. (ciclo while)



```
x = 0
while True:
    print(x)
    if x == 10:
        break
    x += 2
```

#### Ahora vamos a ver la magia de la automatizacion

##### este ciclo while imprime números del 0 al 19 y luego reinicia x a 0 automáticamente. Esto se logra usando la expresión x = (x + 1) % 20, que incrementa x en 1 en cada iteración y luego utiliza el operador módulo % para reiniciarlo a 0 cuando alcanza 20. El bucle while se ejecuta indefinidamente, ya que x nunca supera 19

```
x = 0
while x < 20:
    print(x)
    x = (x+1) % 20
```
